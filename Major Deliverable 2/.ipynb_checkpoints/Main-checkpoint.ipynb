# Naive Algorithm

https://www.geeksforgeeks.org/naive-algorithm-for-pattern-searching/?ref=rp

```python
def naive(string, pattern):
    global counter
    n = len(string)
    m = len(pattern)
    
    for i in range(n - m + 1):
        j = 0
        
        while j < m:
            counter += 1
            
            if string[i+j] != pattern[j]:
                break
                
            j += 1
 
        if j == m:
            found = True
            # print("Pattern found at index ", i)
```

# Rabin-Karp Algorithm

https://www.geeksforgeeks.org/rabin-karp-algorithm-for-pattern-searching/?ref=rp

```python
def rabinKarp(string, pattern):
    global counter
    n = len(string)
    m = len(pattern)
    i = 0
    j = 0
    tHash = 0
    pHash = 0
    h = 1
    d = 256
    prime = 101

    for i in range(m - 1):
        h = (h * d) % prime

    for i in range(m):
        tHash = (d * tHash + ord(string[i])) % prime
        pHash = (d * pHash + ord(pattern[i])) % prime

    for i in range(n - m + 1):
        if pHash == tHash:
            for j in range(m):
                counter += 1
                
                if string[i+j] != pattern[j]:
                    break
                else:
                    j += 1

            if j == m:
                found = True
                # print("Pattern found at index ", i)

        if i < n - m:
            tHash = (d * (tHash - ord(string[i]) * h) + ord(string[i+m])) % prime

            if tHash < 0:
                tHash = tHash + prime
```

# Knuth Morris Pratt (KMP) Algorithm

https://www.geeksforgeeks.org/kmp-algorithm-for-pattern-searching/

```python
def kmp(string, pattern):
    global counter
    n = len(string)
    m = len(pattern)

    lps = [0] * m
    j = 0
    computeLPS(pattern, m, lps)

    i = 0
    while i < n:
        if pattern[j] == string[i]:
            i += 1
            j += 1

        if j == m:
            counter += 1
            found = True
            # print("Found pattern at index ", i - j)
            j = lps[j-1]
        elif i < n and pattern[j] != string[i]:
            if j != 0:
                counter += 1
                j = lps[j-1]
            else:
                i += 1

def computeLPS(pattern, m, lps):
    len = 0
    lps[0]
    i = 1

    while i < m:
        if pattern[i] == pattern[len]:
            len += 1
            lps[i] = len
            i += 1
        else:
            if len != 0:
                len = lps[len-1]
            else:
                lps[i] = 0
                i += 1
```

# Growth Chart: Random String

```python
def getRandomString(num):
    return ''.join(random.choice(strlib.ascii_uppercase) for i in range(num))
```

```python
nums = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
naive_vals = []
rabinKarp_vals = []
kmp_vals = []

for num in nums:
    string = getRandomString(num)
    randNum = random.randint(0, num - 4)
    pattern = string[randNum:randNum+4]
    
    counter = 0
    naive(string, pattern)
    naive_vals.append(counter)
    counter = 0
    rabinKarp(string, pattern)
    rabinKarp_vals.append(counter)
    counter = 0
    kmp(string, pattern)
    kmp_vals.append(counter)
    
fig, ax = plt.subplots()
ax.plot(nums, naive_vals, label='Naive')
ax.plot(nums, rabinKarp_vals, label='Rabin-Karp')
ax.plot(nums, kmp_vals, label='KMP')
ax.legend(loc='upper right')
plt.title('Pattern Searching – Random String')
plt.xlabel('String Length')
plt.ylabel('Frequency Count')
plt.savefig('chart1.jpg')
plt.show()
```

# Growth Chart: Uni-Character String

```python
def getUniCharString(num):
    return 'A' * num
```

```python
nums = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
naive_vals = []
rabinKarp_vals = []
kmp_vals = []

for num in nums:
    string = getUniCharString(num)
    pattern = string[0:4]
    
    counter = 0
    naive(string, pattern)
    naive_vals.append(counter)
    counter = 0
    rabinKarp(string, pattern)
    rabinKarp_vals.append(counter)
    counter = 0
    kmp(string, pattern)
    kmp_vals.append(counter)
    
fig, ax = plt.subplots()
ax.plot(nums, naive_vals, label='Naive')
ax.plot(nums, rabinKarp_vals, label='Rabin-Karp')
ax.plot(nums, kmp_vals, label='KMP')
ax.legend(loc='upper right')
plt.title('Pattern Searching – Uni-Character String')
plt.xlabel('String Length')
plt.ylabel('Frequency Count')
plt.savefig('chart2.jpg')
plt.show()
```

# Growth Chart: AAAAB String

```python
def getABString(num):
    return ('A' * (num - 1)) + 'B' 
```

```python
nums = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
naive_vals = []
rabinKarp_vals = []
kmp_vals = []

for num in nums:
    string = getABString(num)
    pattern = string[num-4:num]
    
    counter = 0
    naive(string, pattern)
    naive_vals.append(counter)
    counter = 0
    rabinKarp(string, pattern)
    rabinKarp_vals.append(counter)
    counter = 0
    kmp(string, pattern)
    kmp_vals.append(counter)
    
fig, ax = plt.subplots()
ax.plot(nums, naive_vals, label='Naive')
ax.plot(nums, rabinKarp_vals, label='Rabin-Karp')
ax.plot(nums, kmp_vals, label='KMP')
ax.legend(loc='upper right')
plt.title('Pattern Searching – AAAAB String')
plt.xlabel('String Length')
plt.ylabel('Frequency Count')
plt.savefig('chart3.jpg')
plt.show()
```

# Growth Chart: BAAAA String

```python
def getBAString(num):
    return 'B' + ('A' * (num - 1))
```

```python
nums = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
naive_vals = []
rabinKarp_vals = []
kmp_vals = []

for num in nums:
    string = getBAString(num)
    pattern = string[0:5]
    
    counter = 0
    naive(string, pattern)
    naive_vals.append(counter)
    counter = 0
    rabinKarp(string, pattern)
    rabinKarp_vals.append(counter)
    counter = 0
    kmp(string, pattern)
    kmp_vals.append(counter)
    
fig, ax = plt.subplots()
ax.plot(nums, naive_vals, label='Naive')
ax.plot(nums, rabinKarp_vals, label='Rabin-Karp')
ax.plot(nums, kmp_vals, label='KMP')
ax.legend(loc='upper right')
plt.title('Pattern Searching – BAAAA String')
plt.xlabel('String Length')
plt.ylabel('Frequency Count')
plt.savefig('chart4.jpg')
plt.show()
```
